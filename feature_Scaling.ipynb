# feature Scaling

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

### here clearly we saw the difference between the values of column Total_bill and Tips . so we try to bring on the same scale
##### hence we use feature scaling

1.   Standarization
2.   Normalisation



```python
# standardisation
# manually ( calculating the zscore )

mean=df['total_bill'].mean()
std=df['total_bill'].std()
mean,std
```

```python
# now we calculating the standard data
std_data=[]
for i in df['total_bill']:
  std_data.append((i-mean)/std)
```

```python
std_data
```

```python
# check the distribution before and after the apply of standardisation
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.distplot(df['total_bill'])
plt.subplot(1,2,2)
sns.distplot(std_data)
```

```python
# here we clearly saw, the their is no difference between the two distribution . hence standardisation does not impact the distribution.thats why it is optional
```

### now we do standarisation using some library

```python
from sklearn.preprocessing import StandardScaler
std_scaler=StandardScaler()
```

```python
data_std_scaler=std_scaler.fit_transform(df[['total_bill']])
```

```python
print(data_std_scaler)
```

```python
#now plot the distribution graph
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.distplot(df['total_bill'])
plt.subplot(1,2,2)
sns.distplot(data_std_scaler)
```

```python
# this is all about the standarisation
```

## Normalisation

```python
#first we do manually
# min-max scalling (Xmin, Xmax)
Xmin=df['total_bill'].min()
Xmax=df['total_bill'].max()

normal_data=[]
for i in df['total_bill']:
  normal_data.append((i-Xmin)/(Xmax-Xmin))
```

```python
print(normal_data)
```

```python
# now saw the distribution plot before and after min-max scalling
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.distplot(df['total_bill'])
plt.subplot(1,2,2)
sns.distplot(normal_data)
```

```python
# no change in the distribution of data before and after of min-max scalling (normalisation )
```

```python
# we do normalisation (min-max scaling using library )
from sklearn.preprocessing import MinMaxScaler
min_max_scaler=MinMaxScaler()
```

```python
minmaxscaler_data=min_max_scaler.fit_transform(df[['total_bill']])
```

```python
print(minmaxscaler_data)
```
