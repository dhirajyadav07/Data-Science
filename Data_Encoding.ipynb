```python
# Label Encoding
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load the dataset "tips " from seaborn library
df=sns.load_dataset("tips")
df.head()
```

```python
# now we goona done label encoding over two column sex and smoker
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
```

```python
# The LabelEncoder is designed to work on one-dimensional data, i.e., a single column at a time
df['sex']=label_encoder.fit_transform(df['sex'])
df['smoker']=label_encoder.fit_transform(df['smoker'])
```

```python
df.head()
```

### 2.Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[['Thur', 'Fri', 'Sat', 'Sun']])
```

```python
df['day'].value_counts()
```

```python
df['day']=oe.fit_transform(df[['day']])
```
