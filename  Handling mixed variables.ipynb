# Handling MIxed Variables

## Handling mixed variables, which can contain both numeric and categorical information, often requires careful processing to extract and clean the data effectively. Let's break down the approach for handling a mixed variable, using a DataFrame where the Ticket column has both numeric and categorical components.

```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv('titanic_mixed_variable.csv')
```

```python
df.sample(5)
```

```python
fig=df['number'].value_counts().plot.bar()
fig.set_title("Passenger travell with")
```

```python
#extract numerical part
df['number_numeric']=pd.to_numeric(df['number'],errors='coerce',downcast='integer')
```

```python
#extract categorical part
df['number_categorical']=np.where(df['number_numeric'].isnull(),df['number'],np.nan)
```

```python
df.head()
```

### now we are going to handle the columns {cabin and Ticket}

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['cabin_num']=df['Cabin'].str.extract('(\d+)')
# we have only a single charecter like C85 so we just extract first char of a sting 

df['cabin_cat']=df['Cabin'].str[0]
# but when we have data like {DSA123,AS34,ZX56} means more than one char 
# then we use regular expression to extract those char from string 
# df['cabin_cat']=df['Cabin'].str.extract('^([A-Z]+)')
```

```python
df.head()
```

```python
df['cabin_cat'].value_counts().plot.bar()
```

```python
df['cabin_num'].value_counts().plot()
```

```python
# Extract Last Part as Number:
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'], errors='coerce', downcast='integer')
# s.split()[-1]: Splits the string by whitespace and selects the last element.
# pd.to_numeric(): Converts the extracted string to a numeric type. errors='coerce' will turn non-numeric values into NaN, and downcast='integer' will convert to the smallest integer type.
```

```python
# Extract First Part as Category:
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
# Handle Numeric Categories:
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan, df['ticket_cat'])
df.head(20)
```
