```python
import numpy as np
```

```python
# now we make a function which doing the same job over each value of size column
def convert_size(item):
  if item[-1]=='M':
    res=float(item.replace(item[-1],''))
    return res*1024
  elif item[-1]=='k':
    res=float(item.replace(item[-1],''))
    return res
  else:
    return str(np.nan)
```

```python
df_copy['Size']=df_copy['Size'].apply(convert_size)
```
